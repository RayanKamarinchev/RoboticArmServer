```python
import numpy as np
import cv2
import glob
save_folder = "./custom_data/cam3_flash_undistorted/"
custom_data = glob.glob("./custom_data/cam3_flash/*.jpg", recursive=True)
len(custom_data)
```

```python
cam_matrix = np.load("../camera_matrix.npy")
dist = np.load("../dist_coeffs.npy") 

img = cv2.imread(custom_data[0])
h, w = img.shape[:2]

undistorted = cv2.undistort(img, cam_matrix, dist)
cv2.imwrite(save_folder + "undistorted.jpg", undistorted)
```

```python
for i, file in enumerate(custom_data):
    img = cv2.imread(file)
    undistorted = cv2.undistort(img, cam_matrix, dist)
    cv2.imwrite(save_folder + f"undistorted_{i}.jpg", undistorted)
```

```python
#renaming
import os
folder = "./top_faces_aruco_data/labels/train/"
custom_data = glob.glob("./top_faces_aruco_data/labels/train/*.txt", recursive=True)

custom_data.sort()

start_counter = 150

for i, old_path in enumerate(custom_data):
    new_name = f"undistorted_{start_counter + i}.txt"
    new_path = os.path.join(folder, new_name)
    
    os.rename(old_path, new_path)

print("Renaming complete.")
```
